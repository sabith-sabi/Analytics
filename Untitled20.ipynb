```python
import pandas as pd
import numpy as np
import matplotlib. pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("C:/Users/DELL/Downloads/Book_1.xlsx")
```

```python
df
```

```python
df.info()
```

```python
team_counts = df['Team'].value_counts()

# Calculate the percentage splitting with respect to the total number of employees
percentage_split = (team_counts / len(df)) * 100

# Print the results
print("Number of employees in each team:")
print(team_counts)
print("\nPercentage splitting with respect to the total number of employees:")
print(percentage_split)
```

```python
team_counts.plot(kind='bar', figsize=(8, 6))
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.title('Employee Distribution by Team')
plt.show()
```

```python
grouped_data = df.groupby('Position')

# Iterate over each group and display the employees
for position, employees in grouped_data:
    print(f"\nEmployees in {position}:")
    print(employees)
```

```python
positions = df['Position'].unique()
employee_data = {}

for position in positions:
    employee_data[position] = df[df['Position'] == position]

# Create a scatter plot for each position
fig, ax = plt.subplots()

for position, data in employee_data.items():
    ax.scatter(data['Salary'], data['Age'], label=position)

ax.set_xlabel('Salary')
ax.set_ylabel('Age')
ax.set_title('Employee Distribution by Position')
ax.legend()
plt.show()
```

```python

# Create a histogram of employee ages
plt.hist(df['Age'], bins=10, edgecolor='black')

# Customize the plot
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.title('Employee Age Distribution')

# Show the plot
plt.show()
```

```python
grouped_data = df.groupby(['Team', 'Position'])['Salary'].sum()

# Find the team and position with the highest and lowest spending
highest_spending = grouped_data.idxmax()
lowest_spending = grouped_data.idxmin()

# Display the results
print("Highest Spending:")
print("Team:", highest_spending[0])
print("Position:", highest_spending[1])
print("Total Salary:", grouped_data[highest_spending])
print()

print("Lowest Spending:")
print("Team:", lowest_spending[0])
print("Position:", lowest_spending[1])
print("Total Salary:", grouped_data[lowest_spending])
```

```python
correlation = df['Age'].corr(df['Salary'])

# Create a scatter plot
plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')

# Display correlation coefficient on the plot
plt.text(0.95, 0.05, f'Correlation: {correlation:.2f}', transform=plt.gca().transAxes, ha='right')

# Show the scatter plot
plt.show()

# Create a heat map of the correlation matrix
correlation_matrix = df[['Age', 'Salary']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Set the title of the heat map
plt.title('Correlation Heat Map: Age vs Salary')

# Show the heat map
plt.show()
```
